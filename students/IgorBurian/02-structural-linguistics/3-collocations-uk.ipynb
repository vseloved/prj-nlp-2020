````python
# https://github.com/lang-uk
# https://github.com/brown-uk/nlp_uk
# https://github.com/stanfordnlp/stanza/blob/master/demo/Stanza_Beginners_Guide.ipynb


# import tokenize_uk
import stanza

stanza.download('uk')

nlp = stanza.Pipeline('uk', processors='tokenize,pos,lemma,mwt,depparse')
````

````python
import re
from collections import defaultdict, Counter

anim = re.compile(r'Animacy=Anim')
book_file = '../../../tasks/02-structural-linguistics/data/tyhrolovy.txt'

with open(book_file) as lines:
    result = []
    
    count = 0
    
    for l in lines:
        if not l.strip(): continue
        
        try:
            doc = nlp(l)
            
            # if count > 25: break   
            # count += 1
            
            for sentence in doc.sentences:
                for w in sentence.words:
                    if w.feats and anim.match(w.feats):
                        for nw in sentence.words:
                            if (nw.pos == 'ADJ' and int(nw.head) == int(w.id)):
                                result.append((nw.lemma, w.lemma))
        except:
            print(l)

stats = sorted(Counter(result).items(), key=lambda x: x[1], reverse=True)

for pair, count in stats:
    print('{}: {} {}'.format(count, pair[0], pair[1]))
````
