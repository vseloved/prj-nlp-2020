# 4. Використання довільних алгоритмів класифікації

```python
from utils import load_data, run_experiment
from vectors import vectorize
import numpy as np

dataset = load_data('cleaned')
```

```python
x = np.vstack(dataset.data.apply(vectorize))
y = np.hstack(dataset.target)
```

## 3.1 LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(C=1, solver='saga', multi_class='multinomial', class_weight='balanced', n_jobs=-1)
run_experiment('lrc', model, x, y)
```

## 3.2 RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(class_weight='balanced', n_jobs=-1)
run_experiment('rfc', model, x, y)
```

## 3.3 SVM 

```python
from sklearn.svm import SVC

model = SVC(kernel='linear', class_weight='balanced')
run_experiment('svc', model, x, y)
```
