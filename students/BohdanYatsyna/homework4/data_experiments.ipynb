```python
import re
import csv
import wikipediaapi

import spacy
nlp = spacy.load("en_core_web_md")


actors = "Brad Pitt"

wiki_wiki = wikipediaapi.Wikipedia('en')

page_py = wiki_wiki.page(actors)
splitted_text = page_py.text.split('\n')
```

```python
count = 0
for line in splitted_text:
    doc = nlp(line)
    count += 1
    print("N:{} TXT:{} \n".format(count,line))
```

```python
from spacy import displacy
doc1 = nlp("Pitt starred in Fight Club (1999) and the heist film Ocean's Eleven (2001), as well as its sequels, Ocean's Twelve (2004) and Ocean's Thirteen (2007).")
displacy.render(doc1, style="dep")
```
