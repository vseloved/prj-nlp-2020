```python
import requests
def get_all_edges(concept):
    offset = 0
    req = requests.get('http://api.conceptnet.io/c/en/' + concept + '?offset=' + str(offset) + '&limit=1000').json()
    all_edges = req
    while len(req['edges']) == 1000:
        offset += 1000
        req = requests.get('http://api.conceptnet.io/c/en/' + concept + '?offset=' + str(offset) + '&limit=1000').json()
        all_edges['edges'] += req['edges']
    return all_edges
```

```python
cat = get_all_edges('cat')
```

```python
len(cat['edges'])
for c in cat['edges']:
    if c['rel'] == 'CapableOf'
        print(c['rel']['label'])
        
# print(c['rel']['label'])
# print(c['start']['label'])
# print(c['end']['label'])
```
