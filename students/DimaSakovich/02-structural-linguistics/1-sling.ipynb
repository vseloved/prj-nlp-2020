```python
import re
import spacy
```

```python
nlp = spacy.load("en_core_web_md")
```

```python
def get_pos_parent(sentence):
    word = re.findall(r"{\w+}", sentence)[0]
    sentence = sentence.replace(word, word[1:-1])
    word = word[1:-1]

    doc = nlp(sentence)
    res = {token.text: (token.pos_, token.head.text)
           if not (token.pos_ == 'VERB' and token.head.text == token.text)
           else ('VERB', "ROOT")
           for token in doc}
    
    return res.get(word)
```

```python
sentences = ["We can {but} hope that everything will be fine.", 
             "It's sad {but} true",
             "Jack brings nothing {but} trouble",
             "Let's do it this {way}!",
             "This is {way} too much!",
             "The prices are going {down}.",
             "Someone pushed him and he fell {down} the stairs.",
             "I’ve been feeling rather {down} lately.",
             "It's not easy to {down} a cup of coffee in one gulp.",
             "Bring a {down} jacket and a pair of gloves, and you'll be fine."]
```

```python
for sentence in sentences:
    pos, parent = get_pos_parent(sentence)
    print(pos, parent, ":", sentence, sep="\t")
```

```python
import stanfordnlp

import warnings
warnings.filterwarnings("ignore")
```

```python
nlp_uk = stanfordnlp.Pipeline(lang='uk')
```

```python
pos_mapping = {'NOUN': 'іменник', 
               'ADP': 'прийменник'}


def get_pos_parent_uk(sentence):
    word = re.findall(r"{\w+}", sentence)[0]
    sentence = sentence.replace(word, word[1:-1])
    word = word[1:-1]

    doc = nlp_uk(sentence)
    res = {token.text: (pos_mapping.get(token.upos, token.upos) , doc.sentences[0].words[token.governor-1].text)
#            if not (token.upos == 'VERB' and token.dependency_relation == "root")
           if not token.dependency_relation == "root"
           else ('дієслово', "ROOT")
           for token in doc.sentences[0].words}
    
    return res.get(word)
```

```python
sentences = ["Рада міністрів Європейського союзу затвердила угоду про спрощений порядок видачі {віз} для України.",
             "Батько Себастьяна {віз} на санях їх театральний гурт до Львова.",
             "А ще дивний елемент інтер’єру – {віз} із продукцією одного з херсонських виробників.",
             "У цю мить {повз} Євгена пролетів останній вагон товарняка.",
             "Кліпнув очима і побачив малого песика, який саме пробігав {повз} у бік села.",
             "Степанко перестав кричати, тільки ламкий стогін {повз} йому із грудей.",
             "Ось присіла на {край} ліжка.",
             "Поставив ту кузню не {край} дороги, як було заведено, а на Красній горі, біля Прадуба.",
             "Розповідаючи про передній {край} лінґвістики, фон Лібіх, як завжди, мислив широко і глобально.",
             "Не {край} мені серце."]
```

```python
for sentence in sentences:
    pos, parent = get_pos_parent_uk(sentence)
#     print(sentence, ":", pos, ",", parent, "\n")
    print(pos, parent, ":", sentence, sep="  ", end="\n\n")
```
