```python
import re
import spacy
from collections import defaultdict
nlp = spacy.load("en_core_web_lg", disable=['textcat', 'ner'])
```

```python
def find_collocations ():
    verbs = {'suppose', 'state', 'announce', 'claim', 'express', 'mention', \
             'affirm', 'tell', 'say', 'speak', 'communicate'}
    # adverbs that can be sentence modifiers, like "Actually, I say it on purpose"
    sent_modifiers = {'actually', 'basically', 'eventually', 'generally', \
                      'simply', 'previously', 'usually'}
    collocations = defaultdict(dict)
    with open('blog2008.txt') as f:
        for line in f:
            tokens = line.split() 
            if any(token.endswith('ly') for token in tokens) \
            and not (tokens[0].lower() in sent_modifiers and tokens[1] == ','):
                processed_sentence = nlp(line)
                for token in processed_sentence:
                    if token.lemma_ in verbs:
                        verb = token.lemma_
                        for child in token.children:
                            if child.dep_ == 'advmod' and child.text.endswith('ly'):
                                adverb = child.text.lower()
                                if adverb not in collocations[verb]:
                                    collocations[verb][adverb] =  1
                                else:
                                    collocations[verb][adverb] += 1
    return collocations 
```

```python
collocations = find_collocations()
```

```python
{k: sorted(v.items(), reverse = True, key = lambda x: x[1])[:10] for k, v in collocations.items()}
```
