```python
with open('../../../../../_test_official-2014.combined-withalt.m2.txt') as f:
    corpus = f.read()
```

```python
def parse_annotation(text):
    parts = text.replace('A ', '').split('|||')
    nums = parts[0].split(' ')
    res = {
        'start': nums[0],
        'end': nums[1],
        'err_tag': parts[1],
        'replacement': parts[2],
        'ann_num': parts[5]
    }
    return res

def get_anns():

    blocks = corpus.split('\n\n')
    res = {}
    for block in blocks:
        group = block.split('\n')
        
        if len(group) < 3:
            continue
        sent = group[0].replace('S ', '')
        res[sent] = {}
        for annot in group[1:]:
            parsed = parse_annotation(annot)
#             if parsed['err_tag'] == 'noop':
#                 continue
            if not res[sent].get(parsed['ann_num']):
                res[sent][parsed['ann_num']] = []
            res[sent][parsed['ann_num']].append(parsed)

    return {sent: anns for sent, anns in res.items() if len(anns.keys()) > 1}
    
anns = get_anns()

def compare_anns(fst, snd):
    matches = 0
    for f in fst:
        for s in snd:
            if f['start'] == s['start'] \
                and f['end'] == s['end'] \
                and f['replacement'] == s['replacement']:
                matches += 1
                print('>> f', f)
                print('>> s', s)
    return matches
    

    
for sent, ann in anns.items():
    print(sent)
    pairwise_list = list(zip(list(ann.values()), list(ann.values())[1:]))

    sent_agr = 0.0

    for fst, snd in pairwise_list:
        matches = compare_anns(fst, snd)
        print('@@', matches/len(fst))
        s_agr = matches/(len(fst) + len(snd))
        sent_agr += s_agr
        print('**', len(fst), len(snd))
        print('$$', agr, s_agr)
    print(sent_agr/len(ann), len(ann))
    
            
    print('\n')
```
