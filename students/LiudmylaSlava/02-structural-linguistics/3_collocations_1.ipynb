```python
import nltk
from nltk.corpus import wordnet as wn
import en_core_web_md
nlp = en_core_web_md.load()
import spacy
```

```python
_list = ["say", "tell", "speak", "claim", "communicate"]
_synonyms = []
for word in _list:
    for syn in wn.synsets(word, pos=wn.VERB): 
        for l in syn.lemmas():
            _synonyms.append(l.name())
print(set(_synonyms))
```

```python
synonyms = _list + ['utter', 'talk', 'pronounce', 'chat', 'convey', 'express']
synonyms
```

```python
text = ['Go to the lord Master, and when no one is near, say quietly to him only one word...', 'She is saying loudly']
def collocate(doc):
    result = {}
    for sen in doc:
        tokens = nlp(sen)
        for token in tokens:     
            if token.pos_=='VERB' and token.lemma_ in synonyms:
                for child in token.children:
                    if child.pos_=="ADV" and child.lemma_.endswith('ly'):
                        res_dict = result.setdefault(token.lemma_, {})
                        res_dict[child.lemma_] = res_dict.get(child.lemma_, 0)+1
    return result
collocate(text)
```

```python
filepath = "../../../tasks/02-structural-linguistics/data/blog2008.txt"
with open(filepath) as fp:
    content = fp.readlines()

content[:5]
```

```python
for word, d in collocate(content).items():
    for adv, count in sorted(d.items(), key=lambda item: item[1], reverse=True):
        print(f"{word} - {adv}: {count}")
```
