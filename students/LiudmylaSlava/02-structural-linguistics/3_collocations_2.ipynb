```python
import stanfordnlp
# stanfordnlp.download('uk')
import pymorphy2
nlp = stanfordnlp.Pipeline(lang='uk')
morph = pymorphy2.MorphAnalyzer(lang='uk')
```

```python
doc = nlp("Шалена кицька з'їла всю сметанку")
doc.sentences[0].print_dependencies()
```

```python
doc.sentences[0].print_tokens()
```

```python
def frequency(doc):
    result = {}
    for sentence in doc.sentences:
        nouns = []
        for token in sentence.words:
            if token.upos=='ADJ' and 'anim' in morph.parse(sentence.words[token.governor-1].text)[0].tag:
                res = result.setdefault(sentence.words[token.governor-1].lemma, {})
                res[token.lemma] = res.get(token.lemma, 0) + 1
    return result
frequency(doc)
```

```python
with open("../../../tasks/02-structural-linguistics/data/tyhrolovy.txt") as f:
    content = f.read()
content[:100]
text = nlp(content)
```

```python
tuples = [(noun, adj, count) for noun, d in frequency(text).items() for adj, count in d.items()]
for noun, adj, count in sorted(tuples, key=lambda item: item[2], reverse=True):
    print(f"{noun} - {adj}: {count}")
```
